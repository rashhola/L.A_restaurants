```python
# Dependencies and Setup

import pandas as pd
import numpy as np
```

```python
# read csv files
inspections_data_path = "Data/COUNTY_OF_LOS_ANGELES_RESTAURANT_AND_MARKET_VIOLATIONS.csv"
inspections_data = pd.read_csv(inspections_data_path)
inspections_data
```

```python
# create a new data frame that has only the serial number and violation code columns

reduced_data = pd.DataFrame(inspections_data)[["SERIAL NUMBER", "VIOLATION  STATUS", "VIOLATION CODE", "POINTS"]].reset_index().assign
reduced_data
```

```python
reduced_data.to_csv("Data/serial_number_and_violation_code_data.csv", index=False)
```

```python
reduced_data.groupby(["SERIAL NUMBER", "VIOLATION  STATUS","VIOLATION CODE"]).count().sort_values(by="POINTS")
```

```python
# create a new data frame that has only the serial number and violation code columns

viocode_viodesc_df = inspections_data[["VIOLATION CODE", "VIOLATION DESCRIPTION"]].drop_duplicates()
viocode_viodesc_df
```

```python
viocode_viodesc_df.to_csv("Data/ violation_code.csv", index=False)
```
